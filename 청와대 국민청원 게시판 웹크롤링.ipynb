```python
no = 1

cno_list = []
title_list = []
people_list = []
category_list = []
start_date_list = []
end_date_list = []
content_list = []
print(str(datetime.today().hour) + ':' + str(datetime.today().minute))

for i in range(0,len(url_list)):
    
    driver.get(url_list[i])

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    time.sleep(2)
    content_all = soup.find('div', 'petitionsView_left_pg')



    #청원등록번호
    c_no = url_list[i].split('/')
    cno_1 = c_no[4].split('?')
    cno_2 = cno_1[0]
    cno_list.append(cno_2)

    #청원제목
    title = content_all.find('h3','petitionsView_title').get_text()
    title_list.append(title)

    #참여인원
    people = content_all.find('h2', 'petitionsView_count').find('span', 'counter').get_text()

    people_list.append(people)

    #카테고리
    cat_all = content_all.find('div','petitionsView_info').find('ul', 'petitionsView_info_list').find_all('li')

    category = cat_all[0].get_text().replace('카테고리','')
    category_list.append(category)

    #청원시작일
    s_date = cat_all[1].get_text().replace('청원시작','')
    start_date_list.append(s_date)

    #청원마감일
    e_date = cat_all[2].get_text().replace('청원마감','')
    end_date_list.append(e_date)

    #내용
    content = content_all.find('div','petitionsView_write').find('div', 'View_write').get_text().replace('\n','')
    if content == '':
        content = content_all.find('div','petitionsView_write').find_all('div', 'View_write')[1].get_text().replace('\n','')
    content = content.replace('\t', "")
    content_list.append(content)
    time.sleep(2)
    
print(str(datetime.today().hour) + ':' + str(datetime.today().minute))
```
