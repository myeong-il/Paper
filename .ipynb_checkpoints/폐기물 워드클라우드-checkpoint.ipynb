```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image
from wordcloud import STOPWORDS
from konlpy.tag import Twitter

import nltk
import sys
import time
import re
import math
import numpy as np
import pandas as pd
import xlwt

import random
import os
```

```python
html = urlopen('https://www.sisajournal.com/news/articleView.html?idxno=184337')

soup = BeautifulSoup(html, 'html.parser')

text = soup.find('div', 'user-snb-wrapper').find_all('p')
```

```python
print(text)
```

```python
text = str(text)
text = re.sub('[a-zA-Z]' , '', text, 0).strip()
text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]', '', text,0).strip()
text = re.sub('<.+?>', '', text, 0).strip()
text = text.replace("있다", "")
print(text)
```

```python
t = Twitter()
text_no = t.nouns(text)
print(text_no)
no = nltk.Text(text_no, name='페트병')
data = no.vocab().most_common(500)
cloud_data = dict(data)
data
```

```python
pet_mask = np.array(Image.open("페트병2.png"))
pet_mask2 = np.array(Image.open("페트병3.jfif"))
pet_mask
```

```python
fig = plt.figure(figsize=(12,12))
plt.imshow(pet_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
font_path = 'c:\\windows\\fonts\\NanumGothic.ttf'
wordcloud = WordCloud(
     font_path = font_path,
    mask = pet_mask,
    background_color='white'
    )

wordcloud = wordcloud.generate_from_frequencies(cloud_data)

fig = plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
fig.savefig('기사1(명사).png')
```
